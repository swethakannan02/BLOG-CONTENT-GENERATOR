```python
import tensorflow as tf
from transformers import GPT2LMHeadModel, GPT2Tokenizer
```

```python
tokenizer = GPT2Tokenizer.from_pretrained("gpt2-large")
```

```python
model = GPT2LMHeadModel.from_pretrained("gpt2-large", pad_token_id=tokenizer.eos_token_id)
```

```python
user_input = input("Enter your blog prompt: ")
```

```python
sentence = user_input
```

```python
input_ids = tokenizer.encode(sentence, return_tensors='pt')
```

```python

input_ids
```

```python
output = model.generate(input_ids, max_length=50, num_beams=5, no_repeat_ngram_size=2, early_stopping=True)
```

```python
output
```

```python
context=tokenizer.decode(output[0], skip_special_tokens=True)
print(context)
```

```python
from IPython.display import HTML, display
context = tokenizer.decode(output[0], skip_special_tokens=True)
html_content = f"""
<!DOCTYPE html>
<html>
<head>
    <title>Generated Blog</title>
</head>
<body>
    <h1>Generated Blog</h1>
    <p>{context}</p>
</body>
</html>
"""
file_name = "generated_blog.html"
with open(file_name, "w") as file:
    file.write(html_content)
html_link = f'<a href="{file_name}" target="_blank">Open Generated Blog</a>'
display(HTML(html_link))
```
